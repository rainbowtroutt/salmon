```python
from google.colab import drive
import csv

drive.mount('/content/drive')

file_path = '/content/drive/My Drive/apart2024.csv'

with open(file_path, newline='', encoding='cp949') as csvfile:
    csvreader = csv.reader(csvfile)

    header = next(csvreader)
    print("Header:", header)

    data = []
    number = []
    number2 = []

    for row in csvreader:
        data.append(row)

answer = input("어떤 구의 부동산 평균값을 알고 싶나요:" )
for row in data:
    if answer in row[2]:
        number.append(int(row[11]))

if number:
    a = len(number)
    average = sum(number) / a
    print(average)
else:
    print("해당 구의 데이터를 찾을 수 없습니다.")

answer2 = input("어떤 구의 부동산 평균값을 비교하고 싶나요:" )
for row in data:
    if answer2 in row[2]:
        number2.append(int(row[11]))

if number2:
    a = len(number2)
    average2 = sum(number2) / a
    print(average2)
else:
    print("해당 구의 데이터를 찾을 수 없습니다.")

```
